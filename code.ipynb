```python
def consommateur():
    while True:
        print(selecteur_choix_client)
        choix = input()
        #Affichage des differents plats
        if choix == "1":
            for element in menu:
                print(f"{element["id"]}: {element["nom"]} | Ingrédients: {", ".join(element["ingredients"])} | Prix: {element["prix"]} € | Allergènes: {", ".join(element["allergenes"])}") 

        #Creation d'une nouvelle commande
        elif choix == "2":
            global nb_commande
            ids = [str(plat["id"]) for plat in menu]
            commande = []
            choix_menu = ""
            while choix_menu != "0":
                print("Quel(s) plat(s) souhaitez-vous commander ?\n(0 pour stopper la commande)")
                choix_menu = input()
                if choix_menu == "0":
                    break     
                elif choix_menu in ids:
                    print("Excellent choix")
                    commande.append(int(choix_menu))
                else:
                    print("Veuillez choisir un plat present dans le menu.")
            if commande:
                total = [menu[plat-1]["prix"] for plat in commande]
                
                nouvelle_commande = {"id_commande": nb_commande, "heure": datetime.now().strftime("%H:%M"), "status": "en préparation", "contenu": commande, "total": sum(total), "timestamp": datetime.now()}
                
                file_attente.enfiler(nouvelle_commande)
                
                print(f"Votre commande est la n'{nb_commande} et a ete envoyer en cuisine !.")
                nb_commande+=1
                

        #Visionner une commande
        elif choix == "3":
            file_actuelle = file_attente.tout_afficher()
            ids_commande = [str(com["id_commande"]) for com in file_actuelle]

            print("Veuillez saisir le numero de votre commande.\n(0 pour retourner en arriere)")
            numero_commande = input()

            if numero_commande in ids_commande:
                for cmd in file_actuelle:
                    if cmd["id_commande"] == int(numero_commande):
                        liste_plat = [menu[plat-1]["nom"] for plat in cmd["contenu"]]
                        print(f"Commande n'{cmd["id_commande"]} fait a {cmd["heure"]}: Contient: {", ".join(liste_plat)} | Status: {cmd["status"]}") 
            else:
                print("Cette commande n'existe pas.")

        #Annuler une commande
        elif choix == "4":
            print("Entrez le numéro de la commande à annuler :")
            numero_commande = input()
            commande_trouve = False
            for cmd in file_attente.tout_afficher():
                if cmd["id_commande"] == int(numero_commande):
                    commande_trouve = True
                    if cmd["status"] != "en préparation":
                        print("Votre commande est déjà envoyée ou bien annulée.")
                    else:
                        delta = datetime.now() - cmd["timestamp"]
                        if delta.total_seconds() <= 120:
                            print("Vous devez attendre au moins 2 minutes avant de pouvoir annuler votre commande. Nos chefs travaillent durs pour confectionner le meilleur des plats en un temps record.")
                        else:
                            choix_annule = input("Voulez-vous annuler votre commande ? (Oui/Non)")
                            if choix_annule.lower() == "oui":
                                cmd["status"] = "annulé"
                                print("Votre commande a ete annulée avec succes. Nous esperons vous revoir bientot.")
                        break
            if not commande_trouve:
                print("Le numero de cette commande n'existe pas.")
        
        #Retour a la selection de role
        elif choix == "5":
            break
        #Renvoie au selecteur si le choix est invalide
        else:
            print("Ce choix est invalide. Veuillez réessayer.")
```
