```python
#Initialisation de fonctions

#Permet de séparer le texte en output de jupiter et la barre d'input en haut de l'éditeur
#Cette fonction est juste un wrapper
def texteShow(texte1, texte2=">"):
    print(texte1)
    return input(texte2)
```

```python
def consommateur():
    while True:

        #Affiche l'UI des choix du consommateur et récupère la saisie
        choix = texteShow(selecteur_choix_conso, "Choix possibles : 1 / 2 / 3 / 4 / 5")
        
        #Affichage du menu
        if choix == "1":
            for element in menu:
                #Affiche les détails de chaque plats de façon esthétique
                print(f"{element["id"]}: {element["nom"]} | Ingrédients: {", ".join(element["ingredients"])} | Prix: {element["prix"]} € | Allergènes: {", ".join(element["allergenes"])}") 

        #Création d'une nouvelle commande
        elif choix == "2":
            #Variables globales pour les statistiques
            global nb_commande, nb_total_commandes

            #Récupération de tous les id pour comparer avec le choix utilisateur
            ids = [str(plat["id"]) for plat in menu]
            commande = []

            #Boucle pour prendre les commandes de plusieurs plats
            while True:
                choix_menu = texteShow("Quel(s) plat(s) souhaitez-vous commander ?\n", "0 pour stopper la commande")
                if choix_menu == "0":
                    break     
                elif choix_menu in ids:
                    print("Excellent choix")
                    commande.append(int(choix_menu))
                else:
                    print("Veuillez choisir un plat présent dans le menu.")
            
            #Si l'utilisateur a commandé quelque chose, on met la commande en préparation
            if commande:
                prix_id = {plat["id"]: plat["prix"] for plat in menu}
                total = [prix_id[i] for i in commande]
                
                nouvelle_commande = {"id_commande": nb_commande,
                                     "heure": datetime.now().strftime("%H:%M"), #Affiche l'heure sous forme 00:00
                                     "status": "en préparation",
                                     "contenu": commande, #Liste des ids commandés
                                     "total": sum(total), 
                                     "timestamp": datetime.now()} #Sert à calculer les temps pour l'annulation, cette valeur n'est jamais affichée
                
                #Ajoute à la file d'attente globale
                file_attente.enfiler(nouvelle_commande)
                nb_total_commandes += 1

                print(f"Votre commande est la n°{nb_commande} et a été envoyée en cuisine !")
                nb_commande+=1
                

        #Visionner le statut d'une commande
        elif choix == "3":

            #Créer une copie du contenu de la file de commande
            file_actuelle = file_attente.tout_afficher()
            #Récupère tous les ids de la file
            ids_commande = [str(com["id_commande"]) for com in file_actuelle]

            numero_commande = texteShow("Veuillez saisir le numéro de votre commande.")

            if numero_commande in ids_commande:
                for cmd in file_actuelle:
                    if cmd["id_commande"] == int(numero_commande):
                        #Récupération du noms des plats par rapport à l'id de la liste 'contenu'
                        liste_plat = [menu[plat-1]["nom"] for plat in cmd["contenu"]]
                        print(f"Commande n°{cmd["id_commande"]} fait a {cmd["heure"]}: Contient: {", ".join(liste_plat)} | Status: {cmd["status"]}") 
            else:
                print("Cette commande n'existe pas.")

        #Annuler une commande
        elif choix == "4":
            numero_commande = texteShow("Entrez le numéro de la commande à annuler :")
            commande_trouve = False

            #Parcourt la file pour trouver la commande
            for cmd in file_attente.tout_afficher():
                if cmd["id_commande"] == int(numero_commande):
                    commande_trouve = True

                    #Vérifie si la commande doit être préparée
                    if cmd["status"] != "en préparation":
                        print("Votre commande est déjà envoyée ou bien annulée.")
                    
                    else:
                        #Vérifie le délai écoulé depuis la commande
                        delta = datetime.now() - cmd["timestamp"]

                        #Si il y moins de 2min alors l'utilisateur ne peux pas annuler sa commande
                        if delta.total_seconds() <= 120:
                            print("Vous devez attendre au moins 2 minutes avant de pouvoir annuler votre commande. Nos chefs travaillent dur pour confectionner le meilleur des plats en un temps record.")
                        else:
                            print("Voulez-vous annuler votre commande ?")
                            choix_annule = input("Choix possibles : Oui / Non")
                            if choix_annule.lower() == "oui":
                                cmd["status"] = "annulée"
                                print("Votre commande a été annulée avec succès. Nous espérons vous revoir bientôt.")
                        break #Permet de sortir prématurément de la boucle for vu que l'on a déjà trouvé notre commande
            if not commande_trouve:
                print("Le numéro de cette commande n'existe pas.")
        
        #Retour a la selection de role
        elif choix == "5":
            break
        else:
            print("Ce choix est invalide. Veuillez réessayer.")
```

```python
def restaurateur():
    while True:

        #Affiche l'UI des choix du consommateur et récupère la saisie
        choix_resto = texteShow(selecteur_choix_resto, "Choix possibles : 1 / 2 / 3 / 4 / 5 / 6 / 7")

        #Afficher les commandes en cours
        if choix_resto == "1":
            file = file_attente
            if file.est_vide():
                print("Aucune commande en attente.")
                return
            print("Commandes en attente :")
            
            file_restant = file.longueur()
            #Parcourt la file en entier pour l'afficher
            while file_restant != 0:
                cmd = file.defiler()
                print(f"- n°{cmd['id_commande']} commandé à {cmd['heure']}, statut {cmd['status']}, total {cmd['total']} €)")
                
                #Met l'élément à sa place pour éviter de le supprimer ou modifier son ordre
                file.enfiler(cmd)
                file_restant -= 1

        #Traiter les commandes en FIFO
        elif choix_resto == "2":
            #Variables globales utilisées pour les statistiques
            global nb_commandes_envoyees, nb_commandes_annulees
            global chiffre_affaires, pertes_annulation

            file = file_attente
            if file.est_vide():
                print("Aucune commande en attente.")
                return
            
            #Retire la commande de la file pour la traiter
            cmd = file.defiler()
            print(f"Traitement de la commande n°{cmd['id_commande']}  de {cmd['heure']}")

                
            if cmd["status"] == "annulée":
                nb_commandes_annulees += 1
                pertes_annulation += cmd["total"]
                print("Commande annulée (demande du client prise en compte).")
            else:
                cmd["status"] = "envoyée"
                nb_commandes_envoyees += 1
                chiffre_affaires += cmd["total"]
                print("Commande envoyée au client.")

        #Ajout d'un nouveau menu
        elif choix_resto == "3":
            print("=== Ajout d'un nouveau plat ===")
            
            nom = texteShow("Nom du plat : ")
            print(nom)
            
            prix_str = texteShow("Prix du plat : ")
            try:
                prix = float(prix_str)
                print(prix_str)
            except ValueError:
                print("Prix invalide, le plat n'est pas ajouté.")
                return
            
            categ = texteShow("Catégorie : ", "Entrée / Plat / Dessert ")

            #Gestion des listes allergènes et ingrédients
            allergenes_str = texteShow("Allergènes : ", "Séparées par des virgules, vide si aucune")
            if allergenes_str.strip() == "":
                allergenes = []
            else:
                allergenes = [a.strip() for a in allergenes_str.split(",")]
            print(allergenes)
            
            print("Ingrédients : ")
            ingredients_str = input("Séparés par des virgules")
            if ingredients_str.strip() == "":
                ingredients = []
            else:
                ingredients = [i.strip() for i in ingredients_str.split(",")]
            print(ingredients)

            #Géneration automatique d'un id unique
            if menu:
                nouvel_id = max(p["id"] for p in menu) + 1
            else:
                nouvel_id = 1

            nouveau_plat = {
                "id": nouvel_id,
                "nom": nom,
                "prix": prix,
                "allergenes": allergenes,
                "categ": categ,
                "ingredients": ingredients
            }
            menu.append(nouveau_plat)
            print(f"Plat ajouté avec l'id {nouvel_id}.")

        #Modification d'un menu existant
        elif choix_resto == "4":
            print("=== Modification d'un plat ===")

            id_str = texteShow("Id du plat à modifier : ")
            try:
                id_plat = int(id_str)
            except ValueError:
                print("Saisie incorrecte.")
                return
            
            #Recherche le plat dans la liste du menu
            plat_liste = [element for element in menu if element["id"] == int(id_plat)]
            if plat_liste == []:
                print("Plat introuvable.")
                return
            
            plat = plat_liste.pop(0)
            
            print(f"Plat actuel : {plat['nom']} | {plat['prix']} €")

            #Change les paramètres uniquement si l'utilisateur le décide de mettre un autre élément
            #Nom du plat
            nouveau_nom = texteShow("Nouveau nom : ", "Vide pour garder le même.")
            if nouveau_nom.strip() != "":
                plat["nom"] = nouveau_nom

            #Prix du plat
            nouveau_prix_str = texteShow("Nouveau prix : ", "Vide pour garder le même.")
            if nouveau_prix_str.strip() != "":
                try:
                    nouveau_prix = float(nouveau_prix_str)
                    plat["prix"] = nouveau_prix
                except ValueError:
                    print("Prix invalide, pas de changement effectué.")

            #Liste des allèrgenes (0 pour supprimer, rien pour garder et nouveau élément pour changer)
            nouvelles_allergenes_str = texteShow("Nouvelles allergènes : ", "Séparées par des virgules, vide si aucun changement, 0 pour effacer l'actuelle.")
            if nouvelles_allergenes_str.strip() != "" and nouvelles_allergenes_str !="0":
                plat["allergenes"] = [a.strip() for a in nouvelles_allergenes_str.split(",")]
            elif nouvelles_allergenes_str == "0":
                plat["allergenes"] = []
                print("Allergènes actuels effacés avec succès")
            
            #Liste des ingrédients (choix de saisie identique au allergènes)
            nouveaux_ingredients_str = texteShow("Ingrédients : ", "Séparés par des virgules, vide si aucun changement.")
            if nouveaux_ingredients_str != "":
                plat["ingredients"] = [i.strip() for i in nouveaux_ingredients_str.split(",")]
            print("Plat modifié.")

        #Suppresion d'un menu existant
        elif choix_resto == "5":
            print("=== Suppression d'un plat ===")
            id_str = texteShow("Id du plat à supprimer : ")
            try:
                id_plat = int(id_str)
            except ValueError:
                print("Saisie invalide.")
                return

            index_a_supprimer = None
            #On cherche aussi l'id pour le supprimer avec 'del'
            for i, plat in enumerate(menu):
                if plat["id"] == id_plat:
                    index_a_supprimer = i
                    break

            if index_a_supprimer is None:
                print("Plat introuvable.")
            else:
                nom_plat = menu[index_a_supprimer]["nom"]
                del menu[index_a_supprimer]
                print(f"Plat '{nom_plat}' supprimé.")

        #Statistiques totales
        elif choix_resto == "6":
            print("===== Statistiques =====")
            print(f"Nombre total de commandes : {nb_total_commandes}")
            print(f"Nombre de commandes envoyées : {nb_commandes_envoyees}")
            print(f"Nombre de commandes annulées : {nb_commandes_annulees}")
            print(f"Chiffre d'affaires : {round(chiffre_affaires, 2)} €")
            print(f"Pertes à cause des annulations : {round(pertes_annulation, 2)} €")
            print("=========================")

        #Retour au sélécteur de rôle
        elif choix_resto == "7":
            break
        else:
            print("Ce choix est invalide. Veuillez réessayer.")
```
