```python
def consommateur():
    while True:
        print(selecteur_choix_client)
        choix = input()
        #Affichage des differents plats
        if choix == "1":
            for element in menu:
                print(f"{element["id"]}: {element["nom"]} | Ingrédients: {", ".join(element["ingredients"])} | Prix: {element["prix"]} € | Allergènes: {", ".join(element["allergenes"])}") 

        #Creation d'une nouvelle commande
        elif choix == "2":
            global nb_commande
            ids = [str(plat["id"]) for plat in menu]
            commande = []
            choix_menu = ""
            while choix_menu != "0":
                print("Quel(s) plat(s) souhaitez-vous commander ?\n(0 pour stopper la commande)")
                choix_menu = input()
                if choix_menu == "0":
                    break     
                elif choix_menu in ids:
                    print("Excellent choix")
                    commande.append(int(choix_menu))
                else:
                    print("Veuillez choisir un plat present dans le menu.")
            if commande:
                total = [menu[plat-1]["prix"] for plat in commande]
                
                nouvelle_commande = {"id_commande": nb_commande, "heure": datetime.now().strftime("%H:%M"), "status": "en préparation", "contenu": commande, "total": sum(total), "timestamp": datetime.now()}
                
                file_attente.enfiler(nouvelle_commande)
                
                print(f"Votre commande est la n'{nb_commande} et a ete envoyer en cuisine !.")
                nb_commande+=1
                

        #Visionner une commande
        elif choix == "3":
            file_actuelle = file_attente.tout_afficher()
            ids_commande = [str(com["id_commande"]) for com in file_actuelle]

            print("Veuillez saisir le numero de votre commande.\n(0 pour retourner en arriere)")
            numero_commande = input()

            if numero_commande in ids_commande:
                for cmd in file_actuelle:
                    if cmd["id_commande"] == int(numero_commande):
                        liste_plat = [menu[plat-1]["nom"] for plat in cmd["contenu"]]
                        print(f"Commande n'{cmd["id_commande"]} fait a {cmd["heure"]}: Contient: {", ".join(liste_plat)} | Status: {cmd["status"]}") 
            else:
                print("Cette commande n'existe pas.")

        #Annuler une commande
        elif choix == "4":
            print("Entrez le numéro de la commande à annuler :")
            numero_commande = input()
            commande_trouve = False
            for cmd in file_attente.tout_afficher():
                if cmd["id_commande"] == int(numero_commande):
                    commande_trouve = True
                    if cmd["status"] != "en préparation":
                        print("Votre commande est déjà envoyée ou bien annulée.")
                    else:
                        delta = datetime.now() - cmd["timestamp"]
                        if delta.total_seconds() <= 120:
                            print("Vous devez attendre au moins 2 minutes avant de pouvoir annuler votre commande. Nos chefs travaillent durs pour confectionner le meilleur des plats en un temps record.")
                        else:
                            choix_annule = input("Voulez-vous annuler votre commande ? (Oui/Non)")
                            if choix_annule.lower() == "oui":
                                cmd["status"] = "annulée"
                                print("Votre commande a ete annulée avec succes. Nous esperons vous revoir bientôt.")
                        break
            if not commande_trouve:
                print("Le numero de cette commande n'existe pas.")
        
        #Retour a la selection de role
        elif choix == "5":
            break
        #Renvoie au selecteur si le choix est invalide
        else:
            print("Ce choix est invalide. Veuillez réessayer.")
```

```python
def restaurateur():
    while True:
        print(selecteur_choix_resto)
        choix_resto = input()

        #Afficher les commandes en cours
        if choix_resto == "1":
            file = file_attente
            if file.est_vide():
                print("Aucune commande en attente.")
                return

            print("Commandes en attente :")
            file_restant = file.longueur()
            while file_restant != 0:
                cmd = file.defiler()
                print(f"- n°{cmd['id_commande']} (heure {cmd['heure']}, statut {cmd['status']}, total {cmd['total']} €)")
                file.enfiler(cmd)
                file_restant -= 1

        #Traiter les commandes une par une du plus ancien au plus nouveau
        elif choix_resto == "2":
            global nb_commandes_envoyees, nb_commandes_annulees
            global chiffre_affaires, pertes_annulation

            file = file_attente
            if file.est_vide():
                print("Aucune commande en attente.")
                return
            
            cmd = file.defiler()
            print(f"Traitement de la commande n°{cmd['id_commande']} (heure {cmd['heure']})")

                
            if cmd["status"] == "annulée":
                file_annule.enfiler(cmd)
                nb_commandes_annulees += 1
                pertes_annulation += cmd["total"]
                print("Commande annulée (demande du client prise en compte).")
            else:
                cmd["status"] = "envoyée"
                nb_commandes_envoyees += 1
                chiffre_affaires += cmd["total"]
                print("Commande envoyée au client.")

        #Ajout d'un nouveau menu
        elif choix_resto == "3":
            print("=== Ajout d'un nouveau plat ===")
            nom = input("Nom du plat : ")
            prix_str = input("Prix du plat : ")
            try:
                prix = float(prix_str)
            except ValueError:
                print("Prix invalide, le plat n'est pas ajouté.")
                return
            categ = input("Catégorie (Entrée / Plat / Dessert) : ")
            allergenes_str = input("Allergènes (séparés par des virgules, vide si aucun) : ")
            if allergenes_str.strip() == "":
                allergenes = []
            else:
                allergenes = [a.strip() for a in allergenes_str.split(",")]
            ingredients_str = input("Ingrédients (séparés par des virgules) : ")
            ingredients = [i.strip() for i in ingredients_str.split(",")]

            # on calcule un nouvel id : on prend le plus grand id + 1
            if menu:
                nouvel_id = max(p["id"] for p in menu) + 1
            else:
                nouvel_id = 1

            nouveau_plat = {
                "id": nouvel_id,
                "nom": nom,
                "prix": prix,
                "allergenes": allergenes,
                "categ": categ,
                "ingredients": ingredients
            }
            menu.append(nouveau_plat)
            print(f"Plat ajouté avec l'id {nouvel_id}.")

        #Modification d'un menu existant
        elif choix_resto == "4":
            print("=== Modification d'un plat ===")
            id_str = input("Id du plat à modifier : ")
            try:
                id_plat = int(id_str)
            except ValueError:
                print("Saisie invalide.")
                return
            
            plat = [element for element in menu if element["id"] == int(id_plat)]
            plat = plat.pop(0)
            if plat is None:
                print("Plat introuvable.")
                return
            print(f"Plat actuel : {plat['nom']} | {plat['prix']} €")

            nouveau_nom = input("Nouveau nom (laisser vide pour garder le même) : ")
            if nouveau_nom.strip() != "":
                plat["nom"] = nouveau_nom

            nouveau_prix_str = input("Nouveau prix (laisser vide pour garder le même) : ")
            if nouveau_prix_str.strip() != "":
                try:
                    nouveau_prix = float(nouveau_prix_str)
                    plat["prix"] = nouveau_prix
                except ValueError:
                    print("Prix invalide, je garde l'ancien prix.")

            nouvelles_allergenes_str = input("Nouvelles allergènes (liste, vide pour garder les mêmes) : ")
            if nouvelles_allergenes_str.strip() != "":
                plat["allergenes"] = [a.strip() for a in nouvelles_allergenes_str.split(",")]
            
            nouveaux_ingredients_str = input("Ingrédients (séparés par des virgules) : ")
            if nouveaux_ingredients_str != "":
                plat["ingredients"] = [i.strip() for i in nouveaux_ingredients_str.split(",")]
            print("Plat modifié.")

        #Suppresion d'un menu existant
        elif choix_resto == "5":
            print("=== Suppression d'un plat ===")
            id_str = input("Id du plat à supprimer : ")
            try:
                id_plat = int(id_str)
            except ValueError:
                print("Saisie invalide.")
                return

            index_a_supprimer = None
            for i, plat in enumerate(menu):
                if plat["id"] == id_plat:
                    index_a_supprimer = i
                    break

            if index_a_supprimer is None:
                print("Plat introuvable.")
            else:
                nom_plat = menu[index_a_supprimer]["nom"]
                del menu[index_a_supprimer]
                print(f"Plat '{nom_plat}' supprimé.")

        #Statistiques totales du restaurant
        elif choix_resto == "6":
            print("===== Statistiques =====")
            print("Nombre total de commandes :", file_attente.longueur())
            print("Nombre de commandes envoyées :", nb_commandes_envoyees)
            print("Nombre de commandes annulées :", nb_commandes_annulees)
            print("Chiffre d'affaires :", round(chiffre_affaires, 2), "€")
            print("Pertes à cause des annulations :", round(pertes_annulation, 2), "€")
            #print("Nombre de commandes dans la pile des annulées :", len(pile_annulees))
            print("=========================")

        #Retour au sélécteur de rôle
        elif choix_resto == "7":
            break
        else:
            print("Ce choix est invalide. Veuillez réessayer.")
```
